I'll create an interactive demonstration of Python's list comprehension with examples you can run and experiment with:




```python

#%%assistant 1
# Basic list comprehension examples
numbers = [1, 2, 3, 4, 5]

# Example 1: Square each number
squares = [num ** 2 for num in numbers]
print("Original numbers:", numbers)
print("Squared numbers:", squares)

# Example 2: Filter even numbers
even_numbers = [num for num in numbers if num % 2 == 0]
print("\nEven numbers:", even_numbers)

# Interactive part - let's create a list from user input
user_numbers = input("Enter some numbers separated by spaces: ")
numbers_list = [int(num) for num in user_numbers.split()]
print("\nYour numbers:", numbers_list)

# Let's do something with user's numbers
doubled = [num * 2 for num in numbers_list]
print("Your numbers doubled:", doubled)

# Filter numbers greater than 5
greater_than_5 = [num for num in numbers_list if num > 5]
print("Numbers greater than 5:", greater_than_5)

# Create a list of tuples with both original and squared values
number_pairs = [(num, num ** 2) for num in numbers_list]
print("\nNumber pairs (original, squared):", number_pairs)
```
